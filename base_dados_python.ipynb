```python
import pandas as pd

#base de dados muito grande, identificando problema para o computador ler um arquivo grande 
#base de dados do kaggle : https://www.kaggle.com/datasets/ealtman2019/ibm-transactions-for-anti-money-Laundering-aml
#HI-Large_Trans.csv nome real da base no kaggle, alterado para large.csv
table = pd.read_csv("large.csv")
display(table)
```

```python
pd.options.display.float_format = "{:,.2f}".format #codigo de formatação numerica

table = pd.DataFrame() #tabela vai iniciar vazia, após vai ter a primeira tabela resumo 

#lendo pedaço por pedaço da base de dados para não estourar o limite de espaço do computador, 
#lendo peçados de 10.000 por vez (chuncksize - chunck )
for chunk in pd.read_csv("large.csv", chunksize=10000): 
    
    #selecionando colunas da base de dados que o kaggle disponiliza, selecionando apenas as colunas que gostaria de mostrar
    colunas = ["Amount Received", "Receiving Currency", "Amount Paid"] 
    
    #selecionando apenas as colunas citadas acima
    chunk_filtrado = chunk[colunas] 
    
    #agrupando o chunck_filtrado e somando as colunas numericas
    resumo = chunk_filtrado.groupby("Receiving Currency", as_index=False).sum() 
    
    #concatenar a tabela original junto com o resumo que possui o calculo 
    table = pd.concat([table, resumo]) 
```

```python
table = table.groupby("Receiving Currency", as_index=False).sum()
display(table)
```
